### Data Visualization

Comparison of the number of book genres

```python
plt.figure(figsize=(10,8))
sns.countplot(
    data=df,
    x=df['Genre'],
    palette='cool'
)
plt.title("Comparison of the number of book genres", fontsize=20)
```

The 20 best authors based on the number of books

```python
plt.figure(figsize=(20,8))
plt.xticks(rotation=90)
sns.countplot(
    data=df,
    x=df['Author'],
    order=df['Author'].value_counts().index[0:20],
    palette='inferno'
)
plt.title("The 20 best authors based on the number of books", fontsize=20)
```

Relationship between price and number of reviews

```python
plt.figure(figsize=(10,8))
sns.scatterplot(
    data=df,
    x="Price",
    y="Reviews",
    hue="Genre",
    palette="winter"
)
plt.title("Relationship between price and number of reviews", fontsize=20)
```

Distribution Plot of user rating

```python
plt.figure(figsize=(10,8))
sns.countplot(
    x=df['User Rating'],
    palette='winter'
)
plt.title('Distribution Plot of user rating',fontsize=20)
```
